```julia
using JuMP
using CPLEX
using GraphPlot 
using LightGraphs
using Colors
using LinearAlgebra
using SparseArrays
```

```julia
function Make_Question(n)
    G_ori = round.(Symmetric(rand(n,n)),digits=3)
    G_final = G_ori - Diagonal(G_ori)
    return G_final
end
```

```julia
function Draw_Question(Graph)
    g = SimpleGraph(Graph)
    nodelabel = 1:nv(g)
    Edge_collection = collect(edges(g))

    edgelabel =  Vector{Float64}()
    for edge in Edge_collection
        append!(edgelabel, Graph[src(edge),dst(edge)])
    end
    gplot(g,nodelabel=nodelabel,edgelabel=edgelabel)
end
```

```julia
function Solve_problem(Graph,n)
    model = Model(with_optimizer(CPLEX.Optimizer))
    set_silent(model)
    @variable(model, y[i = 1:n], Bin)
    @objective(model, Max, 
    sum(
        sum(
            Graph[i,j]*(y[i]+y[j]-2*y[i]*y[j])
            for j = i:n)
        for i = 1:n)
    )
    optimize!(model)
    result = convert(Array{Int64,1}, value.(y) .+1)
    return result
end
```

```julia
function Draw_answer(Graph,result)
    g = SimpleGraph(Graph)
    nodelabel = 1:nv(g)
    Edge_collection = collect(edges(g))

    edgelabel =  Vector{Float64}()
    for edge in Edge_collection
        append!(edgelabel, Graph[src(edge),dst(edge)])
    end
    nodecolor = [colorant"lightseagreen", colorant"orange"]
    gplot(g,nodelabel=nodelabel,edgelabel=edgelabel, nodefillc = nodecolor[result])
end
```

```julia
n =100
K = Make_Question(n)
Draw_Question(K)
```

```julia
result = Solve_problem(K,n)
Draw_answer(K,result)
```

```julia
Draw_answer(K,result)
```
