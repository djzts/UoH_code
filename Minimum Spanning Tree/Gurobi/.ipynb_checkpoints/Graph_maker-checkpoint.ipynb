```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))


def concat(*mats):
    A = np.array(mats[0])
    for B in mats[1:]:
        A = np.concatenate ((A , np.array(B))) 
    return A
```

```python
def density_graph(*mats):
    
    if len(mats) > 1:
    
        plt.figure(figsize=(14, 6 * np.ceil(len(mats)/2)), dpi=100)

        for i, data_x in enumerate(mats):
            
            i += 1

            locals()["ax_1_%d"%(i)] = plt.subplot(2, int(np.ceil(len(mats)//2)) , i)

            locals()["graph_hist_%d"%(i)] = sns.histplot(data= data_x, bins='auto', color = 'blue', ax = locals()["ax_1_%d"%(i)] ,  label = "Frequency")

            locals()["ax_2_%d"%(i)] = locals()["ax_1_%d"%(i)].twinx()

            locals()["graph_dens_%d"%(i)] = sns.kdeplot(data= data_x, color = 'red', ax = locals()["ax_2_%d"%(i)] , label = "Density")

            locals()["lines_%d"%(i)], locals()["labels_%d"%(i)] = locals()["ax_1_%d"%(i)].get_legend_handles_labels()

            locals()["lines2_%d"%(i)], locals()["labels2_%d"%(i)] = locals()["ax_2_%d"%(i)].get_legend_handles_labels()

            locals()["ax_1_%d"%(i)].yaxis.set_major_locator(MaxNLocator(integer=True))

            locals()["ax_1_%d"%(i)].legend(locals()["lines_%d"%(i)] + locals()["lines2_%d"%(i)], locals()["labels_%d"%(i)] + locals()["labels2_%d"%(i)], loc=0, fontsize=14)

            locals()["ax_1_%d"%(i)].set_xlim(0.8 * data_x.min(), 1.2 * data_x.max())

            locals()["ax_1_%d"%(i)].set_xlabel("Second(s)", fontsize=12)

            locals()["ax_1_%d"%(i)].set_title('Histogram and Density Graph',fontsize=15)

            plt.setp(locals()["ax_1_%d"%(i)].get_xticklabels(), fontsize=12)

            plt.setp(locals()["ax_1_%d"%(i)].get_yticklabels(), fontsize=12)

            plt.setp(locals()["ax_2_%d"%(i)].get_yticklabels(), fontsize=12)

            locals()["ax_1_%d"%(i)].yaxis.label.set_size(14)

            locals()["ax_2_%d"%(i)].yaxis.label.set_size(14)


        plt.subplots_adjust(wspace = 0.4 )
        
    else:
        
        plt.figure(figsize=(7, 6), dpi=100)
        
        d = mats[0]
        
        ax1 = plt.subplot(111)

        graph_hist = sns.histplot(data= d, bins='auto', color = 'blue', ax = ax1 ,  label = "Frequency")
        ax2 = ax1.twinx()
        graph_dens = sns.kdeplot(data= d, color = 'red', ax = ax2 , label = "Density")

        lines, labels = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()

        ax1.yaxis.set_major_locator(MaxNLocator(integer=True))
        ax1.legend(lines + lines2, labels + labels2, loc=0,fontsize=14)
        ax1.set_xlim(0.8*d.min(),1.2*d.max())
        ax1.set_xlabel("Second(s)", fontsize=12)
        ax1.set_title('Histogram and Density Graph',fontsize=15)
        plt.setp(ax1.get_xticklabels(), fontsize=12)
        plt.setp(ax1.get_yticklabels(), fontsize=12)
        plt.setp(ax2.get_yticklabels(), fontsize=12)
        ax1.yaxis.label.set_size(14)
        ax2.yaxis.label.set_size(14)
```
