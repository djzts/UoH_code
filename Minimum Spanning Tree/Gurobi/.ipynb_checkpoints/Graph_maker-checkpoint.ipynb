```python
#W3H4 Gurobi
f_1 = [0.00950813, 0.00950813, 0.0070076 , 0.02301788, 0.00900841,
       0.07356071, 0.00950623, 0.03252792, 0.02952576, 0.00750732]
f_2 = [0.00900841, 0.0090065 , 0.00750542, 0.02301979, 0.00900841,
       0.07306099, 0.00950813, 0.03152657, 0.02702141, 0.00850868]
f_3 = [0.00800705, 0.00900841, 0.0070076 , 0.02652168, 0.00850487,
       0.07256126, 0.00850868, 0.03152657, 0.02702332, 0.00750732]


#W3H4 Gurobi
d_1 = [0.0285244 , 0.02252007, 0.02301979, 0.06005096, 0.01651382,
       0.02852631, 0.02902412, 0.02752304, 0.05904961, 0.0340271 ]
d_2 = [0.02752304, 0.02301979, 0.02402115, 0.05854988, 0.01501274,
       0.02902412, 0.02902603, 0.02702141, 0.06155396, 0.03453064]
d_3 = [0.02602386, 0.02351952, 0.02351761, 0.05905151, 0.01551247,
       0.03002548, 0.02752304, 0.02602196, 0.06105232, 0.0375309 ]

d = np.concatenate ((np.array(d_1) , np.array(d_2)))
d = np.concatenate ((np.array(d) , np.array(d_3)))
print(d)
```

```python
#fig, (ax1, ax3) = plt.subplots(1, 2)
plt.figure(figsize=(14, 6), dpi=100)
ax1 = plt.subplot(121)

graph_hist = sns.histplot(data= d, bins='auto', color = 'blue', ax = ax1 ,  label = "Frequency")
ax2 = ax1.twinx()
graph_dens = sns.kdeplot(data= d, color = 'red', ax = ax2 , label = "Density")

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.yaxis.set_major_locator(MaxNLocator(integer=True))
ax1.legend(lines + lines2, labels + labels2, loc=0,fontsize=14)
ax1.set_xlim(0,1.1*d.max())
ax1.set_xlabel("Second(s)", fontsize=12)
ax1.set_title('Histogram and Density Graph',fontsize=15)
plt.setp(ax1.get_xticklabels(), fontsize=12)
plt.setp(ax1.get_yticklabels(), fontsize=12)
plt.setp(ax2.get_yticklabels(), fontsize=12)
ax1.yaxis.label.set_size(14)
ax2.yaxis.label.set_size(14)

## graph 2
ax3 = plt.subplot(122)


ax3.boxplot(d)
x_ticks_labels = ['Box plot']

ax3.set_xticklabels(x_ticks_labels, rotation='horizontal', fontsize=18)
ax3.yaxis.tick_right()
ax3.set_ylabel("Second(s)",fontsize=12)
ax3.yaxis.set_label_position("right")
ax3.axes.xaxis.set_visible(False)
ax3.set_title('Box Plot',fontsize=15)
plt.setp(ax3.get_yticklabels(), fontsize=12)
a = 1
```
