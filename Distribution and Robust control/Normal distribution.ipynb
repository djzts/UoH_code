```python
diff = 0
while diff < 6:
    mu = 0
    sigma = 60
    mu_noise = random.uniform(-0.5, 0.5)
    data = random.normal(mu, sigma, size=300)


    mu_fit, std_fit = norm.fit(data)
    
    diff = np.abs(mu_fit - mu)

print(mu_fit,mu)
```

```python

fig, ax = plt.subplots(figsize=(38, 20))

ax.hist(data,bins = int(50), density=True,alpha=0.8)

# Plot the fit PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p_fit = norm.pdf(x, mu_fit, std_fit)
#
ax.plot(x, p_fit, 'k',linestyle='dashed', linewidth=3, label = "Estimated Distribution Curve")

#
ymin, ymax = plt.ylim()
ax.set_ylim(ymin, ymax)

y = np.linspace(ymin, ymax, 1000)
ax.fill_betweenx(y, mu, mu_fit, facecolor='green', label = "Estimation Bias",alpha=1,zorder=2)
#
p = norm.pdf(x, mu, sigma)
ax.plot(x, p, 'b', linewidth=3, label = "True Distribution Curve")

# mu fit
ax.axvline(x=mu_fit, color='purple', linestyle='--', label = "Estimated Mean", linewidth=3,zorder=2)

# mu 
ax.axvline(x=mu, color='brown', label = "True Mean", linewidth=3,zorder=2)



ax.set_xticklabels([])
ax.set_yticklabels([])


#handles, labels = plt.gca().get_legend_handles_labels()
#by_label = OrderedDict(zip(labels, handles))
#ax.legend(by_label.values(), by_label.keys(),fontsize ="xx-large")
 
#————————————————
#版权声明：本文为CSDN博主「grace_cxj」的原创文章，遵循CC 4.0 BY-SA版权协议，转载请附上原文出处链接及本声明。
#原文链接：https://blog.csdn.net/grace_cxj/article/details/80963211
#ax.legend(fontsize ="xx-large")
ax.legend(fontsize = 43)
```
