```python
import pandas as pd
import os
import numpy as np
```

```python
I = 3
J = 3
N = 2
attempt = 3
```

```python
data = {}
for i in range(2,I+1):
    for j in range(2,J+1):
        for m in range(2,N+1):
            for k in range(1,attempt+1):
                #np.save('Example_I%d_J%d_N%d_ex%d/Data_input'%(I,J,N,attempt), Data_input)
                #np.save('Example_I%d_J%d_N%d_ex%d/Data_input_type'%(I,J,N,attempt), Data_input_type )
                #np.save('Example_I%d_J%d_N%d_ex%d/VM_type'%(I,J,N,attempt), VM_type)
                #np.save('Example_I%d_J%d_N%d_ex%d/VM_speed'%(I,J,N,attempt), VM_speed)
                T_max_true = np.load('Example_I%d_J%d_N%d_ex%d/T_max_true.npy'%(i,j,m,k))
                QUBO_time = np.load('Example_I%d_J%d_N%d_ex%d/QUBO_time.npy'%(i,j,m,k))
                opt_time = np.load('Example_I%d_J%d_N%d_ex%d/opt_time.npy'%(i,j,m,k))
                opt_time_mean = np.load('Example_I%d_J%d_N%d_ex%d/opt_time_divide_N.npy'%(i,j,m,k))
                #np.save('Example_I%d_J%d_N%d_ex%d/answer'%(I,J,N,attempt), answer)
                #np.save('Example_I%d_J%d_N%d_ex%d/VM_speed'%(I,J,N,attempt), VM_speed)
                #np.save('Example_I%d_J%d_N%d_ex%d/VM_DP_time'%(I,J,N,attempt), VM_DP_time)
                #OP.to_csv('Example_I%d_J%d_N%d_ex%d/answer_pandas.csv'%(I,J,N,attempt),index=False)
                #np.save('Example_I%d_J%d_N%d_ex%d/P'%(I,J,N,attempt), P)
                QUBO = np.load('Example_I%d_J%d_N%d_ex%d/QUBO.npy'%(i,j,m,k))
                data["I=%d, J=%d, N=%d, %dth try"%(i,j,m,k)] = [T_max_true, opt_time, QUBO_time, opt_time_mean, np.shape(QUBO)]
                print (data["I=%d, J=%d, N=%d, %dth try"%(i,j,m,k)])
```

```python
table = pd.DataFrame.from_dict(data,orient='index',columns=['Little_greedy_algorithm','Objective_time' , 'QUBO_max_time' , 'average_processing_time_per_machine','QUBO_size'])
```
